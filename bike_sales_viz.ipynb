Selecting the data to analyse

```python
# Analysing which of the company's product categories generate most revenue

df1= df.groupby(by='Product Category', as_index=False).agg({'Revenue':'sum'})

import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize = (8,4))
plt.subplot(title="Revenue by Product Category")
sns.barplot(data=df1, x='Product Category', y='Revenue')


plt.show()
```

From the graph above, the bike category has the highest revenue, so I select the data about sales of bikes for further analysis.

```python

Bikes= df.loc[(df['Product Category'] == 'Bikes')]
Bikes
```

```python
Bikes.info()

# No blank data
```

# Descriptive statistics

Measures of location and measures of variability

```python
print(Bikes['Customer Age'].mode())
print(Bikes['Unit Cost'].mode())
print(Bikes['Unit Price'].mode())
```

```python
print(Bikes['Customer Age'].max()-Bikes['Customer Age'].min())
```

```python
# Distribution of customer age

sns.histplot(data=Bikes, x='Customer Age',bins=10, kde=True)
plt.title('Customer age distribution')
```

```python
# Distribution of quantity

sns.histplot(data=Bikes, x='Quantity', kde=True)
plt.title('Quantity distribution')
```

```python
# Distribution of unit price

sns.histplot(data=Bikes, x='Unit Price',bins=20, kde=True)
plt.title('Unit Price distribution')
```

```python

# This line graph shows Pivot’s monthly bike revenue from the beginning of 2015 to mid-2016. 

Revenue= Bikes.groupby(by='Year_Month', as_index=False).agg({'Revenue':'sum'})

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (15,6))
plt.title('Revenue by year and month')

sns.lineplot(data=Revenue, x='Year_Month', y='Revenue')

plt.xticks(rotation=45)
```

```python
# Create multiple box plots to show the age distribution in four countries

plt.figure(figsize=(10,8))
sns.boxplot(data=Bikes, y='Customer Age', x='Country', orient='v')
```

```python
# Create a bar chart showing the income generated by each generation of consumers

import math
Bikes['generation']=Bikes['Customer Age'].apply(lambda x:math.floor(x/10)*10)

plt.figure(figsize = (8,5))
plt.title("Revenue by Customer Generation")
Bikes.groupby('generation')['Revenue'].sum().plot.barh()
```

```python
# Percent of total sales quantity for each sub-category of the bicycle (Total: in all countries)

plt.figure(figsize = (6,4))
plt.title("Quantity by Sub Category")

Bikes.groupby('Sub Category')['Quantity'].sum().plot.pie(autopct="%1.1f%%")
```
